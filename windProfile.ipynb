```python
import numpy as np
import matplotlib as mpb
import matplotlib.pyplot as plt
```

```python
z = np.arange(0.05, 5, 0.001)
u = np.sqrt(0.015)/0.4 * (np.log(z/0.03) - 6*9.81*0.02*z/(265*0.015))
```

```python
fig, ax = plt.subplots()
plt.plot(z, u)

plt.title('Wind profile')
plt.xlabel('z [m]')
plt.ylabel('u(z)')

ax.set_xscale('log')
ax.set_xticks([0.05, 0.1, 0.2, 0.5, 1, 2, 5])
ax.get_xaxis().set_major_formatter(mpb.ticker.ScalarFormatter())
```
