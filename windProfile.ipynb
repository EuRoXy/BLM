```python
import numpy as np
import matplotlib as mpb
import matplotlib.pyplot as plt
```

Assignment 4a)

```python
z = np.arange(0.05, 5, 0.001)
u = np.sqrt(0.015)/0.4 * (np.log(z/0.03) + 6*9.81*0.008*z/(265*np.sqrt((0.015) ** 3)))
```

```python
fig, ax = plt.subplots()
plt.plot(u, z)

plt.title('Wind profile')
plt.ylabel('z [m]')
plt.xlabel('u(z) [m/s]')

ax.set_yscale('log')
ax.set_yticks([0.05, 0.1, 0.2, 0.5, 1, 2, 5])
ax.get_yaxis().set_major_formatter(mpb.ticker.ScalarFormatter())
```

Assignment 3b)

```python
def f(x):
    if (x < 0):
        return (1-15*x) ** (-1/2) / ((1-16*x) ** (-1/4))
    elif (x >= 0):
        return 1
```

```python
x = np.arange(-2.5, 2, 0.01)

y = []
for i in np.arange(len(x)):
    y.append(f(x[i]))

fig, ax = plt.subplots() #figsize=(7,10))
plt.plot(x,y)

plt.title('$Pr_t$(z/l)')
plt.xlabel('z/L')
plt.ylabel('$Pr_t$')

ax.hlines(0, -2.5, 2, colors='k', linestyle='dashed')
ax.vlines(0, -1, 2, colors='k', linestyle='dashed')

ax.set_xticks(np.arange(-2.5, 2.1, 0.5))
#ax.set_yticks([-10, 0, 1, 5])
ax.get_yaxis().set_major_formatter(mpb.ticker.ScalarFormatter())
```
